    TWA is True Wind Angle

    VMG is Velocity Made Good
    AWS is Apparent Wind Speed

    Flat is Flattening of the sails to reduce the drag of the sails

Other terms:

    VTW is the velocity of the true wind.
    VAW is the velocity of the apparent wind.
    BTW is the bearing of the true wind.
    BAW is the bearing of the apparent wind.
    Vboat is the velocity of the boat.

    (r) reach course 

Mathematical Replations:

True wind conditions given Apparent Wind conditions:

$VTW = sqrt((VAW*sin(BAW))^2+(VAW*cos(BAW)-V_{boat})^2)$

$BTW = atan( (VAW*sin(BAE))/(VAW*cos(BAW)-V_{boat})$

Apparent Wind conditions given True Wind conditions:

$VAW = sqrt( (VTW*sin(BTW))^2 + (VTW*cos(BTW) + V_{boat})^2 )$

$BAW = atan( (VTW*sin(BTW)) / (VTW*cos(BTW)+V_{boat}) )$
